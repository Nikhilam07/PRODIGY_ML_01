Load the Data

```python
import pandas as pd

# Load train and test data
train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/train.csv')

# Show shape and preview
print("Train shape:", train.shape)
print("Test shape:", test.shape)
train.head()
```

Select Useful Features

```python
# Select numerical features that are strongly correlated with SalePrice
numerical_features = [
    'GrLivArea',        # Above ground living area
    'GarageArea',       # Garage size
    'TotalBsmtSF',      # Basement size
    '1stFlrSF',         # 1st floor area
    '2ndFlrSF',         # 2nd floor area
    'OverallQual',      # Overall material quality
    'OverallCond'       # Overall condition
]

# Select categorical features with a limited number of categories
categorical_features = [
    'MSZoning',         # Zoning classification
    'Street',           # Street type
    'Neighborhood',     # Neighborhood
    'BldgType',         # Building type
    'HouseStyle',       # Style
    'ExterQual',        # Exterior quality
    'KitchenQual',      # Kitchen quality
    'GarageFinish'      # Garage interior finish
]

# Combine all selected features
selected_features = numerical_features + categorical_features

# Subset train and test
X = train[selected_features]
X_test = test[selected_features]

# Target variable
y = train['SalePrice']
```

Handle Missing Values & Encode Categorical Features

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

Define Preprocessing Pipelines

```python
# For numerical columns: fill missing with mean
numerical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean'))
])

# For categorical columns: fill missing with most frequent, then one-hot encode
categorical_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

Combine Pipelines

```python
# Combine numerical and categorical pipelines
preprocessor = ColumnTransformer(transformers=[
    ('num', numerical_pipeline, numerical_features),
    ('cat', categorical_pipeline, categorical_features)
])
```

Create the Full Modeling Pipeline

```python
from sklearn.linear_model import LinearRegression
# Create the complete pipeline
model = Pipeline(steps=[
    ('preprocess', preprocessor),         # Cleaning + encoding
    ('regressor', LinearRegression())     # Linear regression model
])

```

```python
from sklearn.metrics import mean_squared_error

# Fit the model on training data
model.fit(X, y)

# Predict on training data
train_preds = model.predict(X)

# Evaluate
mse = mean_squared_error(y, train_preds)
rmse = mse ** 0.5
print(f"Training RMSE: {rmse:.2f}")
```

```python
# Predict house prices on the test set
predictions = model.predict(X_test)
```

```python
# Use 'Id' from test set and predicted SalePrice
submission = pd.DataFrame({
    'Id': test['Id'],
    'SalePrice': predictions
})
```

Accuracy, Stats, Graph

```python
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Predict on training data (for evaluation only)
train_preds = model.predict(X)

# Step 2: Calculate RMSE and RÂ²
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y, train_preds)
rmse = mse ** 0.5
r2 = r2_score(y, train_preds)
print(f"âœ… Training RMSE: {rmse:.2f}")
print(f"âœ… RÂ² (Accuracy Score): {r2 * 100:.2f}%")

# Step 3: Predict on test set
test_preds = model.predict(X_test)

# Step 4: Create submission DataFrame
submission = pd.DataFrame({
    'Id': test['Id'],
    'SalePrice': test_preds
})

# Step 5: Save submission
submission.to_csv('/mnt/data/submission.csv', index=False)

# Step 6: Plot predictions
plt.figure(figsize=(12, 6))
plt.plot(submission['Id'], submission['SalePrice'], marker='o', linestyle='-', color='teal')
plt.title("Predicted House Prices by ID")
plt.xlabel("House ID")
plt.ylabel("Predicted Sale Price")
plt.grid(True)
plt.tight_layout()
plt.show()
```
